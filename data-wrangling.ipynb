```python
import pandas as pd
import numpy as np
```

We create a training data and use it for Data wrangling.

Our Dataset is called Data_Sample.

**Load data**

```python
dataset=pd.read_csv('/kaggle/input/mydata-for-learning-data-wrangling/Data_Sample.csv',sep=";")
```

The dataset has **missing values**:

There are three methods for managing missing value cells:

1. Deleting them
2. Using the average for numerical columns
3. Using the mode for categorical columns

Conversion **(?)** To **non** using numpy

```python
dataset.replace("?",np.nan,inplace=True)
dataset
```

All empty cells and  (?)  becomes **NaN** 
We get a view of your data.

**According to the nature of the data a naming is done for the columns**

```python
column_names=['Name','Sex','Age','Height','Weight','Nationality','Education','Income']
dataset.columns=column_names
```

**First we see which features have NaN values**

```python
nan_counts = dataset.isnull().sum()
nan_counts
```

# MOD method for categorical data

Columns : {Education, Sex , Nationality, Income } are categorical, we use  **MOD**

```python
Idxmax_Education = dataset['Education'].value_counts().idxmax()
Idxmax_Sex = dataset['Sex'].value_counts().idxmax()
Idxmax_Nationality = dataset['Nationality'].value_counts().idxmax()
Idxmax_Income = dataset['Income'].value_counts().idxmax()

dataset['Education'].replace(np.nan , Idxmax_Education , inplace=True)
dataset['Sex'].replace(np.nan , Idxmax_Education , inplace=True)
dataset['Nationality'].replace(np.nan , Idxmax_Education , inplace=True)
dataset['Income'].replace(np.nan , Idxmax_Education , inplace=True)
```

# average method for numerical data

Columns:{ Age, Height, Weight} are numerical, we use **Average**

```python
average1 = int(dataset['Age'].astype('float').mean() )
dataset['Age'].replace(np.nan,average1,inplace=True)

average2 = np.round(dataset['Height'].astype('float').mean(),1)
dataset['Height'].replace(np.nan,average2,inplace=True)

average3 = int(dataset['Weight'].astype('float').mean() )
dataset['Weight'].replace(np.nan,average3,inplace=True)
```

Now let's see the dataset, 
we won't have **missing vlaue** anymore

```python
dataset
```

# How to delete missing value

We're loading your data again.

```python
data = pd.read_csv('/kaggle/input/mydata-for-learning-data-wrangling/Data_Sample.csv',sep=";")
data.replace("?",np.nan,inplace=True)

column_name=['Name','Sex','Age','Height','Weight','Nationality','Education','Income']
data.columns=column_name
data.head(1)
```

**Remove Nans and reset indexes**

```python
data.dropna(subset=['Sex'],inplace=True)
data.dropna(subset=['Age'],inplace=True)
data.dropna(subset=['Height'],inplace=True)
data.dropna(subset=['Weight'],inplace=True)
data.dropna(subset=['Nationality'],inplace=True)
data.dropna(subset=['Education'],inplace=True)
data.dropna(subset=['Income'],inplace=True)

data.reset_index(drop=True , inplace=True)
```
